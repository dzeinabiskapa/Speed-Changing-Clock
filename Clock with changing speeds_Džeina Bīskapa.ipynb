```python
import pygame
import datetime
import math

# Konstantes
WIDTH = 800
HEIGHT = 800
CENTER_W, CENTER_H = WIDTH//2, HEIGHT//2
RADIUS = 400
theta = 0
running = True

# InicializÄ“t pygame
pygame.init()
# EkrÄns
screen = pygame.display.set_mode((WIDTH, HEIGHT))
# Pygame pulkstenis 
clock = pygame.time.Clock()
FPS=120

# DefinÄ“tÄs funkcijas
# Stundu cipariÅ†iem formatÄ“Å¡anas funkcija
def numbers(number, size, position):
    font = pygame.font.SysFont("Cambria", size, True, False)
    text = font.render(number, True, (0,0,0))
    text_rect = text.get_rect(center=(position))
    screen.blit(text, text_rect)

# 2D transformÄcija
def rotate(R, ANGLE):
    x = R * math.sin(ANGLE * math.pi / 180)
    y = R * math.cos(ANGLE * math.pi / 180)
    return x + WIDTH / 2, HEIGHT / 2 - y

# IegÅ«stam realtime 
time_now = datetime.datetime.now()
start_second = time_now.second
start_minute = time_now.minute
start_hour = time_now.hour

# SÄkotnÄ“jo leÅ†Ä·u aprÄ“Ä·ins
start_second_ANGLE = start_second * (360 // 60)
start_minute_ANGLE = (start_minute + start_second // 60) * (360 // 60)
start_hour_ANGLE = (start_hour + start_minute // 60 + start_second // 3600) * (360 // 12)

# PulksteÅ†a rÄdÄ«tÄja garumi
second_R = 340
minute_R = 240
hour_R = 150

# InicializÄ“t rotÄ“Å¡anas leÅ†Ä·us theta
theta_second = 0
theta_minute = 0
theta_hour = 0

# RotÄcijas Ätrums
rotation_speed = 1  # Default rotÄcijas Ätrums 
rotation_speed_change = 0.01  # VÄ“rtÄ«ba par kuru lietotÄjs katru reizi palielina rotÄcijas Ätrumu

# Galvenais pygame loop
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
     
    if running:
        # RotÄcijas Ätruma maiÅ†a izmantojot UP vai DOWN pogas datora klaviatÅ«rÄ
        keys = pygame.key.get_pressed()

        if keys[pygame.K_UP]:
            rotation_speed += rotation_speed_change
        elif keys[pygame.K_DOWN]:
            rotation_speed -= rotation_speed_change

        # PulksteÅ†a uzzÄ«mÄ“Å¡ana
        screen.fill((255,255,255))
        pygame.draw.circle(screen, (230,225,255), (CENTER_W, CENTER_H), RADIUS)
        pygame.draw.circle(screen, (0,0,0), (CENTER_W, CENTER_H), RADIUS, 5)
        pygame.draw.circle(screen, (0,0,0), (CENTER_W, CENTER_H), 15)

        # Stundu ciparu uzzÄ«mÄ“Å¡ana pulkstenÄ«
        for number in range(1,13):
            numbers(str(number), 80, rotate(RADIUS - 80, number * 30))

        # MinÅ«Å¡u strÄ«piÅ†u uzzÄ«mÄ“Å¡ana
        for number in range(0, 360, 6):
                if number % 5:
                    pygame.draw.line(screen, (0,0,0), rotate(RADIUS - 15, number),
                                     rotate(RADIUS - 30, number), 2)
                else:
                    pygame.draw.line(screen, (0,0,0), rotate(RADIUS - 15, number),
                                     rotate(RADIUS - 35, number), 6)    

        # seconds
        if theta_second >= 360:
            theta_second = 0
        else:
            theta_second += 0.06 * rotation_speed

        rotated_ANGLE_seconds =  theta_second + start_second_ANGLE
        pygame.draw.line(screen, (255,0,0), (CENTER_W, CENTER_H), rotate(second_R, rotated_ANGLE_seconds), 4)

        # minutes
        if round(rotated_ANGLE_seconds) == 360:
            theta_minute += 0.36 * rotation_speed

        rotated_ANGLE_minutes =  theta_minute + start_minute_ANGLE
        if rotated_ANGLE_minutes >= 360:
            theta_minute = 0
            start_minute_ANGLE = 0
        pygame.draw.line(screen, (0,0,0), (CENTER_W, CENTER_H), rotate(minute_R, rotated_ANGLE_minutes), 10)

        # hours
        if round(rotated_ANGLE_minutes) >= 360:
            theta_hour += 1.8 * rotation_speed

        if theta_hour >= 360:
            theta_hour = 0

        rotated_ANGLE_hours =  theta_hour + start_hour_ANGLE
        pygame.draw.line(screen, (0,0,0), (CENTER_W, CENTER_H), rotate(hour_R, rotated_ANGLE_hours), 14)

        # VizualizÄ“jam rotation speed priekÅ¡ lietotÄja Ä“rtÄ«bas
        screen.blit(pygame.font.SysFont("Cambria", 12).render(f"Rot speed: {rotation_speed} ", True, (0, 0, 0)), (10, 10))

        clock.tick(FPS)
        pygame.display.update()
        
pygame.quit()
```
