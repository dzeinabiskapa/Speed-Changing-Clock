## Teorētiskā daļa  

<hr><b>Uzdevuma nostādne:</b><br><br>

Realizēt animētā analogā pulksteņa lietotni Python vidē, izmantojot lekcijā apskatītās 2D transformācijas (pagriešana):
* Realizēt minūšu un stundu pulksteņu rādītāju animāciju
* Animācijas izpildei jābūt reallaikā, nevis iepriekš ierakstītai
* Lietotājam jābūt iespējai ietekmēt animācijas ātrumu, norādot leņķi ar kuru notiks pulksteņrādītāja pagriešana
* Pulksteņa centrs ir novietots lietotnes loga centrā

Nedrīkst izmantot iebūvētās metodes, kas veic automātisko pagriešanu. Objekta koordināšu izmaiņas obligāti jārealizē ar formulām, kas tika apskatītas lekcijas laikā.

Konsoles programmas (bez lietotāja saskarnes) netiks pieņemtas.


<hr><b> Pulksteņa rotācijas algoritms</b><br><br>

* Vispirms tiek inicializēts Pygame (bez tā, šī biblotēka neies) un pats Pygame pulkstenis, kā arī nostādītas visas konstantes un izveidots pats pulksteņa logs. 
* Tiek izveidotas funkcijas, kas ir tālāk atbildīgas par stundu ciparu formatējumu un par 2D transformāciju.
* 2D transformācijas funkcija balstās uz lekcijās apskatīto - pagriešanu. Pēc lekcijā skatītā 𝑥′ = 𝑟 ∙ cos(𝜑 + 𝜃) un 𝑦′ = 𝑟 ∙ sin(𝜑 + 𝜃), kur 𝜑 ir sākotnējais leņķis un 𝜃 ir leņķis par kuru mēs gribam pagriezt taisni. Mūsu situācijā priekš x būs sin jāizmanto, bet y - cos. x un y pēc tam pārvieto atbilstoši pulksteņa centram. 
* Pirms galvenā cikla mēs nosakam reālo laiku, šīs nevar atrasties galvenajā cikla, jo tad pulksteni nebus iespējams paātrināt vai palēnināt, jo tas būs reālā laika ierobežots.
* Nosakam visu saistīto ar reālo laiku, jo tas būs pulksteņa sākumpunkts. 
* Definējam kāds būs paātrinājuma default iestatījums un par kādu vērtību lietotājs to var mainīt.
* Inicializējam Pygame galveno ciklu (bez tā Pygame nestrādās), tā ir arī atbildīga par loga aizvēršanu. 
* Definējam kā lietotājs var palielināt paātrinājumu. Izsaucam sarakstu ar iespējamām klaviatūras pogām, kuras lietotājs varētu nospiest. Aprakstam, kas notiek, ja tiek nospiests UP vai DOWN.
* Piepildam logu. Viss logs tiek nokrāsots balts, tad tiek uzzīmēti trīss apļi - itkā melnā līnija apkārt lillā aplim, lillā aplis un pulksteņa centrā mazais melnais aplis.
* Uzzīmējam stundu ciparus un minūšu strīpiņas.
* Leņķi priekš pulksteņa rādītājiem tiek mainīti izmantojot if ciklus, kur rotācijas leņi tiek pakāpeniski palielināti līdz tie sasniedz 360 grādu robežu, kur tie tiek atgriezti uz 0, tā, lai programmai nebūtu jācīnās ar bezgalīgiem skaitļiem. 
* Saskaitot rotēšanas leņķi ar sākuma leņķi var aprēķināt jauno rādītāja pozīciju izmantojot sākumā definēto funkciju "rotate".
* Tiek attēlots "rotation speed" vieglākai lietotāja uztverei par tā veiktajām izmaiņām. 
* clock.tick(FPS) nodrošina konstantus kadrus sekundē pulkstenim. 
* pygame.display.update() nodrošina loga atsvaidzināšanu, atjaunošanos. 


<hr><b>Izmantotās Python bibliotēkas</b><br><br>

<b>pygame</b> - Pygame izvēlēts priekš lietotāja saskarnes (UI) izveidnes, jo šī bibloteka labi tiek galā ar 2D animācijām un ar pamata priekšzinašanām saprotama šī uzdevuma risināšanas sfērā. <br><br>
<b>datetime</b> - Izmantota, lai noteiktu reālo laiku, kas ir pamata laiks (base time) priekš pulkstens rādītāju rotācijas.<br><br>
<b>math</b> - Izmantota, lai būtu iespējams izpildīt darbības kā cos un sin, iegūtu pi vērtību aprēķiniem. Šīs darbības ir vajadzīgas, lai aprēķinātu kā pulkstens rādītāji rotēs, kā tiks izvietoti stundas cipariņi un minūšu atzīmes. <br><br>


## <hr>Praktiskā daļa

```python
import pygame
import datetime
import math

# Konstantes
WIDTH = 800
HEIGHT = 800
CENTER_W, CENTER_H = WIDTH//2, HEIGHT//2
RADIUS = 400
theta = 0
running = True

# Inicializēt pygame
pygame.init()
# Ekrāns
screen = pygame.display.set_mode((WIDTH, HEIGHT))
# Pygame pulkstenis 
clock = pygame.time.Clock()
FPS=120

# Definētās funkcijas
# Stundu cipariņiem formatēšanas funkcija
def numbers(number, size, position):
    font = pygame.font.SysFont("Cambria", size, True, False)
    text = font.render(number, True, (0,0,0))
    text_rect = text.get_rect(center=(position))
    screen.blit(text, text_rect)

# 2D transformācija
def rotate(R, ANGLE):
    x = R * math.sin(ANGLE * math.pi / 180)
    y = R * math.cos(ANGLE * math.pi / 180)
    return x + WIDTH / 2, HEIGHT / 2 - y

# Iegūstam realtime 
time_now = datetime.datetime.now()
start_second = time_now.second
start_minute = time_now.minute
start_hour = time_now.hour

# Sākotnējo leņķu aprēķins
start_second_ANGLE = start_second * (360 // 60)
start_minute_ANGLE = (start_minute + start_second // 60) * (360 // 60)
start_hour_ANGLE = (start_hour + start_minute // 60 + start_second // 3600) * (360 // 12)

# Pulksteņa rādītāja garumi
second_R = 340
minute_R = 240
hour_R = 150

# Inicializēt rotēšanas leņķus theta
theta_second = 0
theta_minute = 0
theta_hour = 0

# Rotācijas ātrums
rotation_speed = 1  # Default rotācijas ātrums 
rotation_speed_change = 0.01  # Vērtība par kuru lietotājs katru reizi palielina rotācijas ātrumu

# Galvenais pygame loop
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
     
    if running:
        # Rotācijas ātruma maiņa izmantojot UP vai DOWN pogas datora klaviatūrā
        keys = pygame.key.get_pressed()

        if keys[pygame.K_UP]:
            rotation_speed += rotation_speed_change
        elif keys[pygame.K_DOWN]:
            rotation_speed -= rotation_speed_change

        # Pulksteņa uzzīmēšana
        screen.fill((255,255,255))
        pygame.draw.circle(screen, (230,225,255), (CENTER_W, CENTER_H), RADIUS)
        pygame.draw.circle(screen, (0,0,0), (CENTER_W, CENTER_H), RADIUS, 5)
        pygame.draw.circle(screen, (0,0,0), (CENTER_W, CENTER_H), 15)

        # Stundu ciparu uzzīmēšana pulkstenī
        for number in range(1,13):
            numbers(str(number), 80, rotate(RADIUS - 80, number * 30))

        # Minūšu strīpiņu uzzīmēšana
        for number in range(0, 360, 6):
                if number % 5:
                    pygame.draw.line(screen, (0,0,0), rotate(RADIUS - 15, number),
                                     rotate(RADIUS - 30, number), 2)
                else:
                    pygame.draw.line(screen, (0,0,0), rotate(RADIUS - 15, number),
                                     rotate(RADIUS - 35, number), 6)    

        # seconds
        if theta_second >= 360:
            theta_second = 0
        else:
            theta_second += 0.06 * rotation_speed

        rotated_ANGLE_seconds =  theta_second + start_second_ANGLE
        pygame.draw.line(screen, (255,0,0), (CENTER_W, CENTER_H), rotate(second_R, rotated_ANGLE_seconds), 4)

        # minutes
        if round(rotated_ANGLE_seconds) == 360:
            theta_minute += 0.36 * rotation_speed

        rotated_ANGLE_minutes =  theta_minute + start_minute_ANGLE
        if rotated_ANGLE_minutes >= 360:
            theta_minute = 0
            start_minute_ANGLE = 0
        pygame.draw.line(screen, (0,0,0), (CENTER_W, CENTER_H), rotate(minute_R, rotated_ANGLE_minutes), 10)

        # hours
        if round(rotated_ANGLE_minutes) >= 360:
            theta_hour += 1.8 * rotation_speed

        if theta_hour >= 360:
            theta_hour = 0

        rotated_ANGLE_hours =  theta_hour + start_hour_ANGLE
        pygame.draw.line(screen, (0,0,0), (CENTER_W, CENTER_H), rotate(hour_R, rotated_ANGLE_hours), 14)

        # Vizualizējam rotation speed priekš lietotāja ērtības
        screen.blit(pygame.font.SysFont("Cambria", 12).render(f"Rot speed: {rotation_speed} ", True, (0, 0, 0)), (10, 10))

        clock.tick(FPS)
        pygame.display.update()
        
pygame.quit()
```
